```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []
low = []
count = -1
end = False

for row in data:
    count += 1
    if count >= 6 and end == False:
        listtest = row[0].split('-')
        month = listtest[1]
        day = listtest[2]
        if month == '03' and day == '01':
            high.append(float(row[4]))
            low.append(float(row[3]))
        if row[0] == '2020-12-31':
            end = True
        
plt.title('Visualize Birthday Temperature')
##한국어 출력 못한다
plt.plot(high, label = 'Highest')
plt.plot(low, label = 'Lowest')
plt.legend()
plt.show()
```
