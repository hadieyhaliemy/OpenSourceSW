```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("data.csv")
city1 = df.iloc[5, 3:104].str.replace(',',"").astype(int)
sum = 0 
for x in range(len(df.columns)-3):
    sum += pow((df.iloc[1:26,(3+x)].str.replace(',',"").astype(int) - city1[x]),2)


df['E'] = sum
max = df[df['E']> 0.0].max()
min = df[df['E']> 0.0].min()
city_max = max.iloc[3:104].str.replace(',',"").astype(int)
city_min = min.iloc[3:104].str.replace(',',"").astype(int)

plt.rcParams["font.family"] = "Gulim"
plt.rcParams["font.size"] = 16
plt.rcParams["figure.figsize"] =(20,10)
plt.title("   ")
plt.style.use('ggplot')
plt.plot(city1, label = str(" : " + df.iloc[5, 0]))
plt.plot(city_max, label = str("  : " +max.iloc[0]))
plt.plot(city_min, label = str("  : " +min.iloc[0]))
plt.legend()
plt.show()
```
